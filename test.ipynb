```python
def prefixe(A):
    if is_empty(A):
        return []
    return [root(A)]+prefixe(left_child(A))+prefixe(right_child(A))

def infixe(A):
    if is_empty(A):
        return []
    return infixe(left_child(A))+[root(A)]+infixe(right_child(A))

def suffixe(A):
    if is_empty(A):
        return []
    return suffixe(left_child(A))+suffixe(right_child(A))+[root(A)]
```

```python

A=["+",["*",[4,[],[]],[5,[],[]]],["/",[6,[],[]],[7,[8,[],[]],[9,[],[]]]]]
anim=animate_secuence(A,suffixe(A),"Le parcour suffixe")
```

```python
#Executer plusiuers fois cette commande
next(anim)
```

```python
def parcour_en_largeur(A):
    file=[A]
    parcour=[]
    while file!=[]:
        node=file.pop(0)
        if not is_empty(node):
            parcour.append(root(node))
            file.extend([left_child(node),right_child(node)])
    return parcour
```

```python
anim=animate_secuence(A,parcour_en_largeur(A),"DFS")
```

```python
try:
    b=next(anim)
except:
    pass
b
```
